```python
import pandas as pd
import matplotlib as plt
from matplotlib import pyplot


import numpy as np
import pandas as pd
import math
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression



%matplotlib inline
```

```python
df = pd.read_csv('Amazon-Products.csv')
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.describe(include='all')
```

```python
to_drop = ['Unnamed: 0', 'image', 'link']
df.drop(columns=to_drop, inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df = df[~df['ratings'].str.contains('₹|FREE|Get', na=False)]
```

```python
df['discount_price'] = df['discount_price'].str.replace('₹|,', '', regex=True)
df['actual_price'] = df['actual_price'].str.replace('₹|,', '', regex=True)

# Conversion de la colonne en type float
df['discount_price'] = pd.to_numeric(df['discount_price'])
df['actual_price'] = pd.to_numeric(df['actual_price'])

df.head()
```

```python
df.dropna(subset=['discount_price','actual_price'], axis=0, inplace=True)

# reseting index
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Creation du LabelEncoder
le = LabelEncoder()

# Fit
df['main_category'] = le.fit_transform(df['main_category'])
df['sub_category'] = le.fit_transform(df['sub_category'])

df.head()
```

```python
df['discount_price'] = df['discount_price'].replace(',', '', regex=True).astype(float)
df['actual_price'] = df['actual_price'].replace(',', '', regex=True).astype(float)
df['ratings'] = df['ratings'].astype(float)
df['no_of_ratings'] = df['no_of_ratings'].replace(',', '', regex=True).astype(float)

df = df.dropna(subset=['ratings', 'no_of_ratings', 'discount_price'])
```

```python
X = df[['main_category',	'sub_category', 'ratings', 'no_of_ratings', 'actual_price']]
y = df['discount_price']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
# prompt: df['main_category'].value_counts() ranger en ordre

df['no_of_ratings'].value_counts()
```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

rfc=RandomForestRegressor(
    max_depth= 7
)
```

```python
rfc.fit(X_train, y_train)
```

```python
y_pred_rfc=rfc.predict(X_test)
```

```python
# prompt: print la performance du modele

from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
print('Performance du modèle:')
print('-----------------------')
print('RMSE:', np.sqrt(mean_squared_error(y_test, y_pred_rfc)))
print('R2 Score:', r2_score(y_test, y_pred_rfc))
```

```python
# prompt: print le score, l'accuracy

print("Score:", rfc.score(X_test, y_test))
print("Accuracy:", r2_score(y_test, y_pred_rfc) * 100)
```

# XGBoost

```python
# prompt: faire un xgboost maintenant

import xgboost as xgb

model_xgb = xgb.XGBRegressor(
    objective="reg:squarederror",
    n_estimators=100,
    max_depth=5
)

```

```python
model_xgb.fit(X_train, y_train)
```

```python
y_pred_xgb = model_xgb.predict(X_test)
print("Score:", model_xgb.score(X_test, y_test))
print("Accuracy:", r2_score(y_test, y_pred_xgb) * 100)
```

```python
# prompt: une table qui compare les deu modeles

models = pd.DataFrame({
    'Model': ['Random Forest', 'XGBoost'],
    'Score': [rfc.score(X_test, y_test), model_xgb.score(X_test, y_test)],
    'Accuracy': [r2_score(y_test, y_pred_rfc) * 100, r2_score(y_test, y_pred_xgb) * 100]
})

print(models.to_string())
```

```python
# Select the model with the highest score
best_model = models[models['Score'] == models['Score'].max()]

# Print the best model
print(f"Best model: {best_model['Model'].values[0]}")
```
